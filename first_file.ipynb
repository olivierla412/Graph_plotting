### To change the xlabel to the course we can use sticks() method


```python
### To change the xlabel to the course we can use sticks() method
w =0.4
x = ['cse','ise','ece','eee','mech']
w = 0.2
boys = [200,40,60,240,300]
girls = [200,70,30,120,15]
bar2 = [i+w for i in bar1]              
plt.bar(bar1,boys,label = 'Boys',width=w)
plt.bar(bar2,girls,label='Girls',width=w)
plt.xticks(bar1, x) # To change the xlabel
## Because the location 0,1 to 4 is present in the bar1 variable
plt.xlabel('Course')
plt.ylabel('Student')
plt.title('Student vs Course')
plt.legend()
plt.show()
```

## placing the label in the middle

```python
### To change the xlabel to the course we can use sticks() method
x = ['cse','ise','ece','eee','mech']
w = 0.4
boys = [200,40,60,240,300]
girls = [200,70,30,120,15]
bar2 = [i+w for i in bar1]                
plt.bar(bar1,boys,label = 'Boys',width=w)
plt.bar(bar2,girls,label='Girls',width=w)
plt.xticks(bar1 +w/2, x)  ## placing the xlabel of item in the middle
plt.ylabel('Student')
plt.title('Student vs Course')
plt.legend()
plt.show()
```

```python
w = 0.2
x = ["USA","Russia","GBR","China","Germany"]
## medals

gold = [20,10,6,24,30]
silver = [100,50,30,120,150]
bronze = [200,100,60,240,300]
bar1 = np.arange(len(x))
bar2 = [i + w for i in bar1]
plt.bar(bar1,gold,w, label="Gold")
plt.bar(bar2,silver,w, label="Silver")



plt.xlabel('Countries')
plt.ylabel('Medals')
plt.legend()
plt.title("Medals vs Countries")
plt.show()
```

```python
w = 0.2
x = ["USA","Russia","GBR","China","Germany"]
## medals

gold = [20,10,6,24,30]
silver = [100,50,30,120,150]
bronze = [200,100,60,240,300]
bar1 = np.arange(len(x))
bar2 = [i + w for i in bar1]
bar3 = [i + w for i in bar2]
plt.bar(bar1,gold,w, label="Gold")
plt.bar(bar2,silver,w, label="Silver")
plt.bar(bar3,bronze,w,label='Bronze')



plt.xlabel('Countries')
plt.ylabel('Medals')
plt.xticks(bar1+w, x,color='r')
plt.legend()
plt.title("Medals vs Countries")
plt.show()
```
